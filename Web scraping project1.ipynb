```python
#Import libraries for web scraping.
#BeautifulSoap is used for parsing and navigating HTML content.
#requests is used for making HTTP reuquests to fecth web pages.
import requests
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/List_of_universities_in_South_Africa'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
#Print content of web page in the form of html to check structure and elements

print(soup)
```

```python
#Locate all <table> elements with their content from the parsed HTML

soup.find_all('table')
```

```python
#Select the second table from the list of <table> in the parsed HTML

table=soup.find_all('table')[1]
```

```python
print(table)
```

```python
#Find all <th> elements from the parsed HTML, these are headings of table 2

soup.find_all('th')
```

```python
#Extract the first 8 elements of <th> which are headers of table 2

university_titles=soup.find_all('th')[0:8]
```

```python
print(university_titles)
```

```python
#Extract text from university title headers, stripping extra spaces and storing them in a list
#Print the list of university headers

sa_university_titles=[title.text.strip() for title in university_titles]
print(sa_university_titles)
```

```python
#Import pandas library which is important for data analysis and manipulation

import pandas as pd
```

```python
# Create a DataFrame using the pandas library with columns set to the university titles 
# that were scraped earlier. This will be used to store the data in a structured table.
df = pd.DataFrame(columns=sa_university_titles)

# Displaying the DataFrame to check its structure
df
```

```python
# Extract all the table rows (<tr>) from the selected table and store them in the 'column_data' variable.
# This will capture the rows containing the actual university data (like names, locations, etc.).
column_data = table.find_all('tr')

# Print the 'column_data' to check the extracted rows from the table
print(column_data)
```

```python
# Extract all the table rows (<tr>) from the second table on the webpage.
# This step captures the individual rows which contain the university data.

column_data = table.find_all('tr')
```

```python
# Iterate through each row in the column_data (excluding the header row).
# For each row, extract all the table data cells (<td>) and clean up the text.
# Then, append the cleaned row data to the DataFrame 'df'.

for row in column_data[1:]:
    row_data = row.find_all('td')    # Find all <td> elements (table data cells) in the row
    individual_row_data = [data.text.strip() for data in row_data]    # Clean up the text in each <td>
    length = len(df)    # Get the current length of the DataFrame
    df.loc[length] = individual_row_data    # Add the cleaned data as a new row to the DataFrame
```
